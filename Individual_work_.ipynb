```python
pcs = hotel_dataset.isnull().sum()
df = pd.DataFrame({'column_name':pcs.index, 'null_count':pcs.values})
plt.figure(figsize=(10,7))
ax = sns.barplot(x="column_name",y='null_count',data=df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
plt.show()
```

```python
hotel_dataset[['agent','company']] = hotel_dataset[['agent','company']].fillna(0.0)
hotel_dataset['country'].fillna(hotel_dataset['country'].mode().to_string(), inplace=True)
hotel_dataset['children'].fillna(round(hotel_dataset['children'].mean()), inplace=True)
```

```python
dhoom = hotel_dataset.isnull().sum()
df3 = pd.DataFrame({'column_name':dhoom.index, 'null_count':dhoom.values})
plt.figure(figsize=(10,7))
ax = sns.barplot(x="column_name",y='null_count',data=df3)

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
plt.tight_layout()
plt.show()
```
